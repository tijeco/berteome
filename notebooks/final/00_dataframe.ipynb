```
# export
class modelPredDF():
    def __init__(self, predDict, seq, aas):
        self.predDf = pd.DataFrame.from_dict(predDict, orient = "index", columns = list(aas))
        self.predDf = self.predDf.div(self.predDf.sum(axis=1),axis=0)
        self.predDf.insert(0, "wt",list(seq))
        self.predDf.insert(1, "wtIndex",list(range(1,len(seq)+1)))
        wtScore = self.wtScoreCol()
        self.predDf.insert(2, "wtScore",wtScore)

    def wtScoreCol(self):
        wtScore = []
        for row in self.predDf.to_dict(orient="records"):
	        wt = row["wt"]
	        wtScore.append(row[wt])
        return wtScore
        

        
        
```

# Ensemble predictions

I think the easiest way to ensemble predictions would be to just take the average probability across all 20 residues. I imagine this would be a relatively simple dataframe function.

```
t1 = pd.DataFrame.from_dict({
								"s1":[1,3,5],
								"s2":[2,4,6],
								"s3":[1,2,3]
							}, orient="index", columns = ["A","B","C"]).reset_index()
t2 = pd.DataFrame.from_dict({
								"s1":[3,1,3],
								"s2":[4,2,8],
								"s3":[3,4,5]
							}, orient="index", columns = ["A","B","C"]).reset_index()
```

Alright, so first there needs to be a way to concatenate them and give them a label column.

```
def stackDFs(df1,df1Name, df2, df2Name):
	df1["label"] = df1Name
	df2["label"] = df2Name

	return pd.concat([df1,df2])
```

In all liklihood this could take a dict {dfName:df..} so that an arbitrary number of dataframes could be provided..

```
tBoth = stackDFs(t1,"t1",t2,"t2")
```

```
tBoth.groupby("index").mean()
```

So yeah, that works! We can just throw this into a function, and we should have a way to get a consensus between ESM and prot_bert!
